```python
import pandas as pd
```

```python
df=pd.read_csv("corona_dat.csv")
```

```python
df.head()
```

```python
df2=df[["date","China","Italy","Brazil","Spain","United Kingdom", "US","France","Turkey"]]
```

```python
df2.set_index("date",inplace=True)
```

```python
df2
```

```python
total=df2.cumsum(axis=0)
```

**total=df2.cumsum(axis=0)** is a pandas operation that calculates the cumulative sum of each column in the dataframe df2. **cumsum()** stands for "cumulative sum".

The parameter axis=0 specifies that the operation is performed column-wise, meaning that each cell in a column represents the cumulative sum of all the cells above it, including itself.

```python
total
```

```python
import bar_chart_race as bcr
```

```python
bcr.bar_chart_race(total, filename="covid.mp4",figsize=(8,5),title="Daily Increase Rate of COVID-19 Cases")
```
